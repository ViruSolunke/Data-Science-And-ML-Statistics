```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import normal   # normal function is used to generate the normal distribution data from numpy
sample = normal(loc=50, scale=5, size=1000) # Mean is {loc=50}, Standard deviation is {scale=5}, size of Data is {size=1000} 
```

```python
sample.mean()
```

```python
sample.std()
```

```python
# plot histogram to understand the distribution of data
plt.hist(sample,bins=10)
```

```python
# Calculate sample Mean and sample Std Deviation
sample_mean = sample.mean()
sample_std = sample.std()
```

```python
# fit the Distribution with above parameter
from scipy.stats import norm
dist = norm(sample_mean, sample_std) # create a normal distribution
```

```python
values = np.linspace(sample.min(),sample.max(),100)  # It generates values ​​spaced equally between the start and stop points. ex :- np.linspace(10, 50, 5) Output :- array([10., 20., 30., 40., 50.])
```

```python
sample.max()
```

```python
sample.min()
```

```python
probabilities = [dist.pdf(value) for value in values] # dist stores the object of a norm() and we use pdf() through that which are the library function of a scipy.states 
```

```python
# Plot a histogram and PDF
plt.hist(sample,bins=10,density=True)
plt.plot(values,probabilities)
```
