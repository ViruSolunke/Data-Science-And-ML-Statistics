```python
import pandas as pd
import random

L = []
for i in range(10000):
    L.append(random.randint(1,6))
len(L)
print(L[:5])

''' 2 Dies '''

L2 = []
for i in range(10000):
    a = random.randint(1,6)
    b = random.randint(1,6)

    L2.append( a + b )

len(L2)
print(L2[:5])
```

```python
pmf = (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()
print(pmf)  

''' 2 Dies Probability Mass Function '''

pmf2 = (pd.Series(L2).value_counts()/pd.Series(L2).value_counts().sum()).sort_index()
print(pmf2)
```

```python
pmf.plot(kind='bar')
```

```python
pmf2.plot(kind='bar')
```

```python
''' Cumulative Distribution Function of PMF Which is the sum of all probabilities up to a certain value '''

import numpy as np
np.cumsum(pmf2).plot(kind='bar')
```
